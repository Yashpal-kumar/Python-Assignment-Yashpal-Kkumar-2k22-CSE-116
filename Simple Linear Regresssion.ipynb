```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from pandas.core.common import random_state
from sklearn.linear_model import LinearRegression
```

```python
df_dbt=pd.read_csv("diabetes.csv")
df_dbt.head()
```

```python
df_dbt.describe()
```

```python
plt.title('Age')
sns.histplot(df_dbt['Age'])
plt.show()
```

```python
plt.scatter(df_dbt['Age'], df_dbt['BMI'], color = 'r')
plt.title('Age vs BMI')
plt.xlabel('AGE')
plt.ylabel('BMI')
#plt.box(False)
plt.show()
```

```python
X = df_dbt[['Age']]  
y = df_dbt['BMI'] 
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

```python
y_pred_train = regressor.predict(X_train)
plt.scatter(X_train, y_train, color = 'r')
plt.plot(X_train, y_pred_train, color = 'b')
plt.title('Age Vs BMI (Training Set)')
plt.xlabel('Age')
plt.ylabel('BMI')
plt.legend(['X_train/Pred(y_test)', 'X_train/y_train'], title = 'BMI/Age', loc='best', facecolor='green')
plt.box(False)
plt.show()
```

```python
y_pred_train = regressor.predict(X_train)
plt.scatter(X_test, y_test, color = 'r')
plt.plot(X_train, y_pred_train, color = 'b')
plt.title('Age vs BMI (Test Set)')
plt.xlabel('Age')
plt.ylabel('BMI')
plt.legend(['X_train/Pred(y_test)', 'X_train/y_train'], title = 'Age/BMI', loc='best', facecolor='green')
plt.box(False)
plt.show()
```

```python
print(f'Coefficient: {regressor.coef_}')
print(f'Intercept: {regressor.intercept_}')
```
